<h1>Data Visualization Example with Python
<h2>Exploring the Titanic Dataset

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

<h2>Load data

```python
titanic = sns.load_dataset('titanic')
```

<h2>Examine data

```python
#Examine data type 
titanic.info()
```

```python
#Saw earlier that age and deck do not each have 891 obs, so they are missing observations
#Calculate no. of NaN for each feature
titanic.isnull().sum(axis=0)
```

```python
#First few observations
titanic.head()
```

```python
#Set theme
sns.set_style('whitegrid')

#Set color palette
sns.set_palette('colorblind') 
```

<h2>Distribution Plot

```python
sns.distplot(titanic.age.dropna(),kde=True,rug=True,)
sns.plt.title('Age Distribution of Titanic Passengers')
sns.plt.ylabel('Proportion')
sns.plt.xlabel('Age')
sns.plt.ylim(0,0.035)
sns.plt.xlim(0,100)
sns.plt.yticks(size=12)
sns.plt.xticks(rotation=0,size=12)
sns.plt.show()
```

<h2>Regression Plot

```python
sns.regplot(x="age", y="fare", data=titanic)
sns.plt.show()
```

<h2>Count Plot

```python
sns.countplot(x="sex", hue="class", data=titanic)
sns.plt.show()
```

<h2>Box Plot

```python
sns.boxplot(x='alive',y='age',hue='class',data=titanic)
sns.plt.show()
```

<h2>Violin Plot

```python
sns.violinplot(x='alive',y='age',hue='class',data=titanic,inner=None)
sns.plt.show()
```

<h2>Facet Grid Plots

```python
g = sns.FacetGrid(titanic, row='survived', col='sex')
g.map(sns.distplot, "age")
sns.plt.show()
```

<h2>Heatmap

```python
#Calculate correlations of all the numerical features
corr = titanic.corr() 
#Generate heatmap
sns.heatmap(corr,annot=True) 
sns.plt.show()
```

<h2>Joint Plot

```python
sns.jointplot('age','fare', data=titanic)
sns.plt.show()
```

<h2>Kernel Density Plot

```python
sns.jointplot(titanic.age,titanic.fare[titanic.fare<50], kind='kde') #fare dist has very long tail..
sns.plt.show()
```
